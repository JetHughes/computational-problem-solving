```python
import numpy as np
import pandas as pd
import seaborn as sns
from matplotlib import pyplot as plt
```

```python
data = pd.read_csv("results_linear_inc10.csv")
# Plot Highest vs. Cost
plt.plot(data['Highest'], data['Cost'])

plt.xlabel('Highest safe floor')   # Set the x-axis label
plt.ylabel('Cost')   # Set the y-axis label
plt.title('Cost vs. Highest safe floor for increment of 10')   # Set the plot title
plt.grid(True)   # Show gridlines (optional)

plt.show()   # Display the plot
```

```python
data = pd.read_csv("results.csv")
grouped_data = data.groupby('Increment')
# Create a box plot for each increment level
plt.figure(figsize=(10, 8))
plt.boxplot([group['Cost'] for increment, group in grouped_data], labels=[increment for increment, group in grouped_data])

plt.xlabel('Increment')
plt.ylabel('Cost')
plt.title('Box Plot of Cost for Various Levels of Increment')
plt.grid(True)

plt.show()
```

```python
data = pd.read_csv("fineincrement.csv")
grouped_data = data.groupby('Increment')['Cost'].max()
# Create a line plot with maximum values for each increment
plt.figure(figsize=(10, 8))
plt.plot(grouped_data.index, grouped_data.values)

plt.xlabel('Increment')
plt.ylabel('Cost')
# plt.yscale('log')
plt.title('Line Plot of Maximum Cost for Each Increment')
plt.xticks(np.arange(0,100, 10))
plt.grid()

plt.show()
```

```python
data = pd.read_csv("linearTwo.csv")
grouped_data = data.groupby('Increment')['Cost'].max()
# Create a line plot with maximum values for each increment
plt.figure(figsize=(10, 8))
plt.plot(grouped_data.index, grouped_data.values)

plt.xlabel('Inital Floor')
plt.ylabel('Cost')
# plt.yscale('log')
plt.title('Line Plot of Worst Case vs Inital Floor')
plt.xticks(np.arange(0,100, 10))
plt.grid()

plt.show()
```

```python
data = pd.read_csv("linearTwoCoarse.csv")
grouped_data = data.groupby('Increment')
# Create a box plot for each increment level
plt.figure(figsize=(10, 8))
plt.boxplot([group['Cost'] for increment, group in grouped_data], labels=[increment for increment, group in grouped_data],whis=5)

plt.xlabel('Initial Floor')
plt.ylabel('Cost')
plt.title('Box Plot of Cost for Various Inital Floors')
plt.grid(True)

plt.show()
```

```python
data = pd.read_csv("linearTwoCoarse.csv")
increment = 14

# Filter the data for the desired increment level
filtered_data = data[data['Increment'] == increment]

plt.figure(figsize=(8, 6))
plt.plot(filtered_data['Floor'], filtered_data['Cost'])

plt.xlabel('Highest Safe Floor')
plt.ylabel('Cost')
plt.title('Cost vs Highest Safe Floor with Inital Floor of 14')
plt.grid(True)

plt.show()
```

```python
plt.figure(figsize=(8, 6))

data = pd.read_csv("linearTwoCoarse.csv")
filtered_data = data[data['Increment'] == 14]
plt.plot(filtered_data['Floor'], filtered_data['Cost'],label='New Method')

data = pd.read_csv("fineincrement2.csv")
filtered_data = data[data['Increment'] == 10]
plt.plot(filtered_data['Floor'], filtered_data['Cost'],label='Original Method')



plt.xlabel('Highest safe floor')   # Set the x-axis label
plt.ylabel('Cost')   # Set the y-axis label
plt.legend()
plt.title("Cost of New Method vs Cost of Original Method")
plt.grid(True)   # Show gridlines (optional)

plt.show()   # Display the plot
```
